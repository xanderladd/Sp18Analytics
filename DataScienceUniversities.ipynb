```
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  
import pandas as pd
import numpy as np
import re
import io
import statistics
df = pd.read_csv(io.StringIO(uploaded['timesMergedData.csv'].decode('utf-8')))
df
```

```
### Average amount of credits
dur = np.asarray(df['DURATION'])
dur

contains_credit_arr = []
for i in dur:
  i.lower()
  if 'credit' in i:
    contains_credit_arr = np.append(contains_credit_arr, i)

contains_credit_arr

# get number of credits from string
credit_hours = []
for i in contains_credit_arr:
  credit_hours = np.append(credit_hours, re.findall("(\d+) c", i))
credit_hours = credit_hours.astype(int)

avg = np.average(credit_hours)
print("Average credits: " + str(avg))
```

```
### Value Counts for Pre-reqs

prereq_col = np.asarray(df['PREREQ'])
prereq_col

prereqs = []
for i in prereq_col:
  if 'Not Available' not in i:
    prereqs = np.append(prereqs, i.lower())
prereqs

numCS = 0
numMath = 0
numStats = 0
numBA = 0
numBio = 0

for i in prereqs:
  if 'computer science' in i:
    numCS += 1
  if 'math' in i:
    numMath += 1
  if 'statistics' in i:
    numStats += 1
  if 'bachelor' in i: 
    numBA += 1
  if 'biology' in i:
    numBio += 1
prereqs

print('Number w/ CS prereq: ' + str(numCS))
print('Number w/ Math prereq: ' + str(numMath))
print('Number w/ Stats prereq: ' + str(numStats))
print('Number w/ Bio prereq: ' + str(numBio))
print('Number w/ Bachelor\'s degree prereq: ' + str(numBA))


```

```
# count NaNs in f_m_ratio

count_nan = len(df) - df.F_M_RATIO.count()
proportion_nans = count_nan/len(df)
print("Number of NaNs in F_M_RATIO: " + str(count_nan))
print("Proportion of NaNs in F_M_RATIO: " + str(round(proportion_nans,2)))



```

```
# median, standard deviation in female male ratio

ratio_col = np.asarray(df['F_M_RATIO'].dropna())
#print(ratio_col)

num_Male = []
num_Female = []
for i in ratio_col:
    num_Male = np.append(num_Male, i[:2])
    if i[3] == ":":
      num_Female = np.append(num_Female, i[5:7])
    else:
      num_Female = np.append(num_Female, i[3:5])

num_Male = [float(i) for i in num_Male]
num_Female = [float(i) for i in num_Female]

#print(num_Female)
median_Male = statistics.median(map(float, num_Male))
median_Female = 100 - median_Male

std_Male = np.std(num_Male)
std_Female = np.std(num_Female)

print("Median Male to Female Ratio: " + str(int(median_Male)) + ":" + str(int(median_Female)))
print("Std Deviation for Male to Female Ratio: " + str(round(std_Male,3)) + ":" + str(round(std_Female, 3)))
```

```
# number/percentage of online programs (delivery col) vs. campus programs
df.DELIVERY
valCounts = df.DELIVERY.value_counts()
print(valCounts)
listVals = valCounts.tolist()
total = sum(listVals)

num_Campus = listVals[0] + listVals[7] 
num_Online = listVals[1]
num_Either = listVals[2] + listVals[3] + listVals[5] + listVals[8] + listVals[10]
num_Hybrid = listVals[4] + listVals[9] + listVals[11] + listVals[12] + listVals[13]

print("\nDistribution of on-campus and online programs:")
print("Number on-campus only: " + str(num_Campus) + " --> " + str(round(num_Campus/total, 2)) + "%")
print("Number online only: " + str(num_Online) + " --> " + str(round(num_Online/total, 2)) + "%")
print("Number on-campus OR online: " + str(num_Either) + " --> " + str(round(num_Either/total, 2)) + "%")
print("Number on-campus AND online: " + str(num_Hybrid) + " --> " + str(round(num_Hybrid/total, 2)) + "%")
```
